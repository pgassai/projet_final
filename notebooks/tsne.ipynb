# Librarie

```python
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE
import pandas as pd
import matplotlib.pyplot as plt
```

# Import des données

```python
data = pd.read_csv("../data/city_lifestyle_dataset.csv")
```

```python
data.head()
```

```python
data.describe()
```

### Description rapide des données :

Les variables avec les plus disperser sont la population_density, avg_income et avg_rent.

Les données city_name et country sont des variables qualitatives non utilisable pour la reduction de dimension T-SNE.

# Réduction de la Dimention : T-SNE

```python
tsne = TSNE(n_components=2, random_state=42, perplexity=30)

data_tsne = tsne.fit_transform(data.drop(columns=['city_name', 'country']))
```

```python
pd.DataFrame(data_tsne).head()
```

```python
kmeans = KMeans(n_clusters=5, random_state=42)
data['cluster'] = kmeans.fit_predict(data_tsne)
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(
    data_tsne[:, 0], 
    data_tsne[:, 1],
    c=data['cluster'], 
    alpha=0.6,  s=30,
    cmap="Paired")
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.title('Dimension Reduction via t-SNE')
plt.grid(True)
plt.show()
```

On remarque que les cinq clusters créés à l'aide du K-Means sont bien séparés à part pour le violet qui a deux observations proche d'un autre cluster.

# Export des données réduites

```python
tsne_df=pd.DataFrame(
    data_tsne,
    columns=["TSNE_1","TSNE_2"]
)

tsne_df["city_name"]    =   data["city_name"].values
tsne_df["country"]      =   data["country"].values

tsne_df.to_csv("../outputs/tsne_emb_2d.csv", index=False)
```
